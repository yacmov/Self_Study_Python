# 11. 데이터 정렬(Sorting values)

```python
import pandas as pd
df = pd.read_excel('score.xlsx', index_col='지원번호')
df
```

```python
df.sort_values('키') # 키 기준으로 오름차순 정렬
```

```python
df.sort_values('키', ascending=False)
```

```python
df.sort_values(['수학', '영어']) # t수학, 영어 점수 기준으로 오름차순 
```

```python
df.sort_values(['수학', '영어'], ascending=False) # t수학, 영어 점수 기준으로 내림차순 
```

```python
df.sort_values(['수학', '영어'], ascending=[True, False]) # ascending을 리스트로 보내면 각각에 맞게 오름차순 내림차순으로 설정이 된다.
```

```python
df['키'].sort_values()
```

```python
df['키'].sort_values(ascending=False)
```

```python
df.sort_index()
```

```python
df.sort_index(ascending = False)
```

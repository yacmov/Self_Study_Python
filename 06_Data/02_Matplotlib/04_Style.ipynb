# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windwos
# matplotlib.rcParamsp['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y, linewidth = 5)
```

## 마커

```python
plt.plot(x,y, marker ='o')
```

```python
plt.plot(x,y, marker='o', linestyle='None')
```

```python
plt.plot(x,y, marker='v')
```

```python
plt.plot(x,y, marker='v', markersize =10)
```

```python
plt.plot(x,y, marker='X', markersize=10)
```

>참고 링크  
>https://matplotlib.org/stable/api/markers_api.html

```python
plt.plot(x,y, marker='o', markersize=20, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize=20, markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일

```python
plt.plot(x,y, linestyle=':')
```

```python
plt.plot(x,y, linestyle='--')
```

```python
plt.plot(x,y, linestyle='-.')
```

## 포멧

```python
plt.plot(x,y, 'ro--') # 색깔r, 마커 스타일 o, 선 스타일 --
```

```python
plt.plot(x,y, 'bv:')
```

```python
plt.plot(x,y,'go') # 라인 스타일을 정해주지 않으면 선이 나타나지 않는다.
```

```python
 plt.plot(x,y, color='g', marker='o', linestyle='None') # 위에 코드랑 같은 결과 값을 출력 한다.
```

## 축약어

```python
plt.plot(x,y, marker='o', mfc='r', ms=10, mec='b', ls=':')
# mfc = markerfacecolor
# ms = markersize
# mec = markeredgecolor
# ls = linestyle
```

>참고 링크  
>https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html

## 투명도

```python
plt.plot(x,y, marker='o', mfc='r', ms=10, alpha=0.7) # alpha : 투명도 [0~1]
```

## 그래프 크기

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5), dpi=200) # dots per inch
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor='y')
plt.plot(x,y)
```

```python
plt.figure(facecolor='#a1c3ff')
plt.plot(x,y)
```

# 8. 막대 그래프 (기본)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # Windwos
# matplotlib.rcParamsp['font.family'] = 'AppleGothic' # Mac
matplotlib.rcParams['font.size'] = 15 # 글자 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 한글 폰트 사용시, 마이너스 글자가 깨지는 현상을 해결
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187 ,184] # 키

plt.bar(labels, values)
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187 ,184] # 키

plt.bar(labels, values, color ='r')
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187 ,184] # 키
colors = ['r', 'g', 'b']

plt.bar(labels, values, color =colors, alpha=0.5)
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187 ,184] # 키

plt.bar(labels, values)
plt.ylim(175, 195) # Y 축에 제약을 줄수 있다
```

```python
plt.bar(labels, values, width=0.5)
plt.ylim(175, 195) # Y 축에 제약을 줄수 있다
```

```python
plt.bar(labels, values, width=0.3)
plt.xticks(rotation=45) # X 축의 이름 데이터 각도를 45 도로 설정
plt.yticks(rotation=45) # Y 축의 이름 데이터 각도를 45 도로 설정
plt.ylim(175, 195)
```

```python
labels = ['강백호', '서태웅', '정대만'] # 이름
values = [190, 187 ,184] # 키
ticks = ['1번 학생', '2번 학생', '3번 학생']


plt.bar(labels, values)
plt.xticks(labels, ticks, rotation=90)
```
